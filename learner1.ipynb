```python
#importing required libraries
import pandas as panda 
import math 
from sklearn.metrics import confusion_matrix

#read training  and validation data from given files
train_file = panda.read_csv("/Users/poojaakula/Desktop/Deep_Learning_HW/hw1/train.txt", delimiter=r"\s+", header=None, names=["i_1", "i_2", "i_3"])
valid_file = panda.read_csv("/Users/poojaakula/Desktop/Deep_Learning_HW/hw1/valid.txt", delimiter=r"\s+", header=None, names=["i_1", "i_2", "i_3"])

#initializing the extracted features training data
x_1 = train_file["i_1"]
x_2 = train_file["i_2"]
y = train_file["i_3"]

#initializing the extracted features validation data
test_x1 = valid_file["i_1"]
test_x2 = valid_file["i_2"]
target = valid_file["i_3"]

#initializing weights
w0, w1, w2,   = 1,-1,0

#initializing learning rate
Eta = 0.001

#defining Sigmoid function
def sigmoid(s):
 sig = 1/(1+ math.exp(-s)) 
 return sig

#defining no of iterations and iterating through given train data to get updated weights
iterations= 250
sse = 0
for j in range(iterations):   
    for i in range(len(x_1)):
        yCap = sigmoid(w0 + w1*x_1[i] + w2*x_2[i])
        diff = (y[i]-yCap)
        sse = sse + (diff *diff)
        delta = (y[i]-yCap)*yCap*(1-yCap)
        w0 = w0 + Eta*delta
        w1 = w1 + Eta*delta*x_1[i] 
        w2 = w2 + Eta*delta*x_2[i]
        i+=1

#printing the updated weights and SSE
print("WEIGHTS:\nw0:", w0, "\nw1:", w1, "\nw2:" , w2, "\nSum-of-Squares Error:", sse)

#Computing predicting values using 
pre_values = []
for i in range(len(test_x1)):
    z = w0 + w1 * test_x1[i] + w2 * test_x2[i]
    pre_ouput = sigmoid(z)

    if (pre_ouput > 0.5):
        pre_values.append(1)
    else:
        pre_values.append(0)
#printing the confusion matrix
print("Confusion matrix with the given values is\n", confusion_matrix(target, pre_values)) 

#printing the Accuracy
matrix = confusion_matrix(target, pre_values)
print("Accuracy between the target and predicted values is ",matrix.diagonal()/matrix.sum(axis=1))

#writing the contents to a text file
file = open("/Users/poojaakula/Desktop/Deep_Learning_HW/learner1output.txt", "w")

file.write(str(w0)+" "+str(w1)+" "+str(w2)+ "\n")
file.write(str(sse)+ ":\n")

file.write("\nCS-5001: HW#1" + "\n")
file.write("Programmer: Sree Pooja Akula"+ "\n")

file.write("\nTRAINING:" + "\n")
file.write("Using learning rate eta ="+str(Eta)+ "\n")
file.write("Using "+str(iterations)+" iterations." + "\n")

file.write("\nLEARNED:" + "\n")
file.write("w0 = "+str(w0) + "\n")
file.write("w1 = "+str(w1) + "\n")
file.write("w2 = "+str(w2) + "\n")

file.write("\nVALIDATION:" + "\n")
file.write("Sum-of-Squares Error = "+str(sse) + "\n")

file.close()
```
